```python
from PIL import Image, ImageDraw
import os
import xml.etree.ElementTree as ET
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
from glob import glob

def get_class_names_yaml_logodet(path):
    classes = []
    for folder in glob(path + '/*/', recursive=True):
        for subfolder in glob(folder + '/*/', recursive=True):
            classes.append(subfolder.split('/')[-2])
    return classes

def get_class_names_logodet(path):
    classes = {}
    class_number = 0
    for folder in glob(path + '/*/', recursive=True):
        for subfolder in glob(folder + '/*/', recursive=True):
            class_name = subfolder.split('/')[-2]
            classes[class_name] = class_number
            class_number += 1
    return classes

def get_path_names(path):
    annotations = []
    for folder in glob(path + '/*/', recursive=True):
        for subfolder in glob(folder + '/*/', recursive=True):
            for xml in glob(subfolder + '*.xml'):
                annotations.append(xml)
    return annotations

def get_image_paths(path):
    images = []
    for folder in glob(path + '/*/', recursive=True):
        for subfolder in glob(folder + '/*/', recursive=True):
            for image in glob(subfolder + '*.jpg'):
                images.append(image)
    return images

def extract_info_from_xml(xml_file):
    root = ET.parse(xml_file).getroot()

    info_dict = {}
    info_dict['bboxes'] = []

    for elem in root:
        if elem.tag == "filename":
            info_dict['filename'] = elem.text
        elif elem.tag == "size":
            image_size = [int(subelem.text) for subelem in elem]
            info_dict['image_size'] = tuple(image_size)
        elif elem.tag == "object":
            bbox = {}
            for subelem in elem:
                if subelem.tag == "name":
                    bbox["class"] = subelem.text
                elif subelem.tag == "bndbox":
                    for subsubelem in subelem:
                        bbox[subsubelem.tag] = int(subsubelem.text)
            info_dict['bboxes'].append(bbox)

    return info_dict

def convert_to_yolov5(info_dict, image_path, class_name_to_id_mapping):
    print_buffer = []

    for b in info_dict["bboxes"]:
        try:
            class_id = class_name_to_id_mapping[b["class"]]
            b_center_x = (b["xmin"] + b["xmax"]) / 2
            b_center_y = (b["ymin"] + b["ymax"]) / 2
            b_width = (b["xmax"] - b["xmin"])
            b_height = (b["ymax"] - b["ymin"])

            image_w, image_h = info_dict["image_size"]
            b_center_x /= image_w
            b_center_y /= image_h
            b_width /= image_w
            b_height /= image_h

            print_buffer.append("{} {:.3f} {:.3f} {:.3f} {:.3f}".format(class_id, b_center_x, b_center_y, b_width, b_height))
        except KeyError:
            print("Invalid Class. Must be one from ", class_name_to_id_mapping.keys())

    out_ann = image_path.replace(".jpg", ".txt")
    with open(out_ann, "w") as f:
        f.write("\n".join(print_buffer))

def plot_bounding_box(image, annotation_list, class_id_to_name_mapping):
    w, h = image.size
    plotted_image = ImageDraw.Draw(image)

    transformed_annotations = np.copy(annotation_list)
    transformed_annotations[:, [1, 3]] = annotation_list[:, [1, 3]] * w
    transformed_annotations[:, [2, 4]] = annotation_list[:, [2, 4]] * h

    transformed_annotations[:, 1] = transformed_annotations[:, 1] - (transformed_annotations[:, 3] / 2)
    transformed_annotations[:, 2] = transformed_annotations[:, 2] - (transformed_annotations[:, 4] / 2)
    transformed_annotations[:, 3] = transformed_annotations[:, 1] + transformed_annotations[:, 3]
    transformed_annotations[:, 4] = transformed_annotations[:, 2] + transformed_annotations[:, 4]

    for ann in transformed_annotations:
        obj_cls, x0, y0, x1, y1 = ann
        plotted_image.rectangle(((x0, y0), (x1, y1)))
        plotted_image.text((x0, y0 - 10), class_id_to_name_mapping[int(obj_cls)])

    plt.imshow(np.array(image))
    plt.show()

# Simulate the argparse arguments
class Args:
    pass

opt = Args()
opt.dataset = 'LogoDet-3K'
opt.plot = True
opt.image = 'data/flickr_logos_27_dataset/flickr_logos_27_dataset_images/4771736332.txt'

annotations_path = 'data/flickr_logos_27_dataset/flickr_logos_27_dataset_training_set_annotation.txt'
class_name_to_id_mapping = get_class_names(annotations_path)
class_id_to_name_mapping = dict(zip(class_name_to_id_mapping.values(), class_name_to_id_mapping.keys()))
image_paths = get_image_paths('data/flickr_logos_27_dataset/flickr_logos_27_dataset_images')

for image_path in tqdm(image_paths):
    xml_path = image_path.replace(".jpg", ".xml")
    if os.path.exists(xml_path):
        info_dict = extract_info_from_xml(xml_path)
        convert_to_yolov5(info_dict, image_path, class_name_to_id_mapping)

if opt.plot:
    annotation_file = opt.image
    with open(annotation_file, "r") as file:
        annotation_list = file.read().split("\n")[:-1]
        annotation_list = [x.split(" ") for x in annotation_list]
        annotation_list = [[float(y) for y in x] for x in annotation_list]

    image_file = annotation_file.replace("txt", "jpg")
    assert os.path.exists(image_file)

    image = Image.open(image_file)
    plot_bounding_box(image, annotation_list, class_id_to_name_mapping)
```

```python
from PIL import Image, ImageDraw
import os
import xml.etree.ElementTree as ET
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
from glob import glob

def get_class_names_logodet(path):
    classes = {}
    class_number = 0
    for folder in glob(path + '/*/', recursive=True):
        for subfolder in glob(folder + '/*/', recursive=True):
            class_name = subfolder.split('/')[-2]
            classes[class_name] = class_number
            class_number += 1
    return classes

def get_annotations_logodet(path):
    annotations = []
    for folder in glob(path + '/*/', recursive=True):
        for subfolder in glob(folder + '/*/', recursive=True):
            for image_path in glob(subfolder + '/*.jpg'):
                annotations.append(image_path)
    return annotations

def extract_info_from_xml(xml_file):
    root = ET.parse(xml_file).getroot()

    info_dict = {}

    filename_elem = root.find('filename')
    size_elem = root.find('size')
    object_elems = root.findall('object')

    if filename_elem is not None and size_elem is not None and object_elems:
        info_dict['filename'] = filename_elem.text

        width = int(size_elem.find('width').text)
        height = int(size_elem.find('height').text)
        info_dict['image_size'] = (width, height)

        info_dict['bboxes'] = []
        for obj_elem in object_elems:
            bbox_elem = obj_elem.find('bndbox')
            if bbox_elem is not None:
                bbox = {
                    'class': obj_elem.find('name').text,
                    'xmin': int(bbox_elem.find('xmin').text),
                    'ymin': int(bbox_elem.find('ymin').text),
                    'xmax': int(bbox_elem.find('xmax').text),
                    'ymax': int(bbox_elem.find('ymax').text),
                }
                info_dict['bboxes'].append(bbox)
    else:
        print(f"Error: Missing required elements in {xml_file}")

    return info_dict

def convert_to_yolov5(info_dict, ann, class_name_to_id_mapping):
    print_buffer = []

    for b in info_dict["bboxes"]:
        try:
            class_id = class_name_to_id_mapping[b["class"]]

            b_center_x = (b["xmin"] + b["xmax"]) / 2
            b_center_y = (b["ymin"] + b["ymax"]) / 2
            b_width = (b["xmax"] - b["xmin"])
            b_height = (b["ymax"] - b["ymin"])

            image_w, image_h = info_dict["image_size"]
            b_center_x /= image_w
            b_center_y /= image_h
            b_width /= image_w
            b_height /= image_h

            print_buffer.append("{} {:.3f} {:.3f} {:.3f} {:.3f}".format(class_id, b_center_x, b_center_y, b_width,
                                                                        b_height))
        except KeyError:
            print("Invalid Class. Must be one from ", class_name_to_id_mapping.keys())

    out_ann = ann.split('.')[0]
    out_ann = out_ann.split('/')
    del out_ann[-1]
    out_ann = '/'.join(out_ann)
    save_file_name = os.path.join(out_ann, info_dict["filename"].replace("jpg", "txt"))

    print("\n".join(print_buffer), file=open(save_file_name, "w"))

def plot_bounding_box(image, annotation_list, class_id_to_name_mapping):
    w, h = image.size
    plotted_image = ImageDraw.Draw(image)

    for b in annotation_list:
        obj_cls = b["class"]
        x_min, y_min, x_max, y_max = b["xmin"], b["ymin"], b["xmax"], b["ymax"]

        x0 = x_min * w
        y0 = y_min * h
        x1 = x_max * w
        y1 = y_max * h

        plotted_image.rectangle(((x0, y0), (x1, y1)))
        plotted_image.text((x0, y0 - 10), obj_cls)

    plt.imshow(np.array(image))
    plt.show()

# Simulate the argparse arguments
class Args:
    pass

opt = Args()
opt.dataset = 'LogoDet-3K'
opt.plot = True
opt.image = r'C:\Users\envy\Downloads\LogoDet-3K\Transportation\zamboni\16.jpg'  # Change the extension to .jpg

annotations_path = 'data/LogoDet-3K'
class_name_to_id_mapping = get_class_names_logodet(annotations_path)
class_id_to_name_mapping = {v: k for k, v in class_name_to_id_mapping.items()}
image_paths = get_annotations_logodet(annotations_path)

print(image_paths)

for image_path in tqdm(image_paths):
    xml_path = image_path.replace(".jpg", ".xml")
    if os.path.exists(xml_path):
        info_dict = extract_info_from_xml(xml_path)
        convert_to_yolov5(info_dict, image_path, class_name_to_id_mapping)

if opt.plot:
    annotation_file = opt.image.replace("jpg", "xml")
    info_dict = extract_info_from_xml(annotation_file)

    image_file = opt.image
    assert os.path.exists(image_file)

    image = Image.open(image_file)
    plot_bounding_box(image, info_dict['bboxes'], class_id_to_name_mapping)
```
